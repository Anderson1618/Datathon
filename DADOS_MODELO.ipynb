```python
import pandas as pd

# Carregar o arquivo original
file_path = '/content/BD_final.csv.csv'
df = pd.read_csv(file_path, sep=';')

# Criação de dataframes separados por ano
df_2020 = df[['ID_ALUNO', 'ANO_INGRESSO', 'INDE_2020', 'PEDRA_2020', 'PONTO_VIRADA_2020']].dropna()
df_2020['ano'] = 2020

df_2021 = df[['ID_ALUNO', 'ANO_INGRESSO', 'INDE_2021', 'PEDRA_2021', 'PONTO_VIRADA_2021']].dropna()
df_2021['ano'] = 2021

df_2022 = df[['ID_ALUNO', 'ANO_INGRESSO', 'INDE_2022', 'PEDRA_2022', 'PONTO_VIRADA_2022']].dropna()
df_2022['ano'] = 2022

# Renomear colunas para padronizar
df_2020.columns = ['ID_ALUNO', 'ANO_INGRESSO', 'INDE', 'PEDRA', 'PONTO_VIRADA', 'ano']
df_2021.columns = ['ID_ALUNO', 'ANO_INGRESSO', 'INDE', 'PEDRA', 'PONTO_VIRADA', 'ano']
df_2022.columns = ['ID_ALUNO', 'ANO_INGRESSO', 'INDE', 'PEDRA', 'PONTO_VIRADA', 'ano']

# Concatenar os dataframes
df_combined = pd.concat([df_2020, df_2021, df_2022])

# Salvar o novo dataframe com a coluna 'ano'
output_path_with_year = '/content/sample_data/dados_combinados_com_ano_correto.csv'
df_combined.to_csv(output_path_with_year, index=False)

# Exibir as primeiras linhas do dataframe atualizado
df_combined.head()
```
