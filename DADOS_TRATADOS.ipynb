```python
!pip install pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import regexp_replace, col, when, length, round, pow, lit, year, current_date, to_date, floor, split
from pyspark.sql.types import IntegerType, FloatType, DoubleType, DecimalType, LongType
```

```python
#DATA SET
spark = SparkSession.builder.getOrCreate() # Create a SparkSession object
df_pm = spark.read.option("sep", ";").csv('/content/PEDE_PASSOS_DATASET_FIAP.csv', header=True, inferSchema=True)
```

```python
df_pm.show()
```

```python
df_pm.printSchema()
```

```python
colunas_notas = ['NOTA_PORT_2022', 'NOTA_MAT_2022', 'NOTA_ING_2022']
```

```python
#AJUSTE DE BD NOTAS 0 a 10
for coluna in colunas_notas:
    df_pm = df_pm.withColumn(coluna,
                       when(col(coluna).contains(".") &
                            (length(regexp_replace(col(coluna), "[^.]", "")) > 1),
                            round(regexp_replace(col(coluna), "\\.", "").cast("double") / 1000000000000000, 1))
                       .otherwise(round(col(coluna).cast("double"), 1)))
```

```python
df_pm.show()
```

```python
colunas_INDE = ["INDE_2020", 'IAA_2020', 'IEG_2020', 'IPS_2020', 'IDA_2020', 'IPP_2020', 'IPV_2020', "INDE_2021", 'INDE_2022', 'IAA_2022', 'IEG_2022', 'IPS_2022', 'IDA_2022', 'IPP_2022', 'IPV_2022']
```

```python
for coluna in colunas_INDE:
    df_pm = df_pm.withColumn(coluna, regexp_replace(col(coluna), "\\.", ""))

df_pm.show()
```

```python
df_pm.printSchema()
```

```python
for coluna in colunas_INDE:
    df_pm = df_pm.withColumn(coluna, col(coluna).cast(LongType()))
```

```python
df_pm.show()
```

```python
#AJUSTE DE INDICES DO BD
for coluna in colunas_INDE:
    df_pm = df_pm.withColumn(f"{coluna}_resultado_operacao",
                        round(col(coluna) / pow(10, length(col(coluna).cast("string")) - 1), 2))

df_pm.show()
```

```python
df_pm = df_pm.drop("INDE_2020", 'IAA_2020', 'IEG_2020', 'IPS_2020', 'IDA_2020', 'IPP_2020', 'IPV_2020', "INDE_2021", 'INDE_2022', 'IAA_2022', 'IEG_2022', 'IPS_2022', 'IDA_2022', 'IPP_2022', 'IPV_2022')

df_pm.show()
```

```python
#RENOMEANDO AS COLUNAS
df_pm = df_pm.withColumnRenamed("INDE_2020_resultado_operacao", "INDE_2020").withColumnRenamed("IAA_2020_resultado_operacao", "IAA_2020").withColumnRenamed("IEG_2020_resultado_operacao", "IEG_2020").withColumnRenamed("IPS_2020_resultado_operacao", "IPS_2020").withColumnRenamed("IDA_2020_resultado_operacao", "IDA_2020").withColumnRenamed("IPP_2020_resultado_operacao", "IPP_2020").withColumnRenamed("IPV_2020_resultado_operacao", "IPV_2020").withColumnRenamed("INDE_2021_resultado_operacao", "INDE_2021").withColumnRenamed("INDE_2022_resultado_operacao", "INDE_2022").withColumnRenamed("IAA_2022_resultado_operacao", "IAA_2022").withColumnRenamed("IEG_2022_resultado_operacao", "IEG_2022").withColumnRenamed("IPS_2022_resultado_operacao", "IPS_2022").withColumnRenamed("IDA_2022_resultado_operacao", "IDA_2022").withColumnRenamed("IPP_2022_resultado_operacao", "IPP_2022").withColumnRenamed("IPV_2022_resultado_operacao", "IPV_2022").withColumnRenamed("IDA_2022_resultado_operacao", "IDA_2022").withColumnRenamed("IPP_2022_resultado_operacao", "IPP_2022").withColumnRenamed("NOME", "ID_ALUNO").withColumnRenamed("ANO_INGRESSO_2022", "ANO_INGRESSO")
```

```python
ano_atual = 2020
```

```python
#JOIN DE INFORMAÇÕES
df_pm = df_pm.withColumn("ANO_INGRESSO",
                   when(col("ANO_INGRESSO").isNull(),
                        lit(ano_atual) - col("ANOS_PM_2020"))
                   .otherwise(col("ANO_INGRESSO")))

df_pm.show()
```

```python
#LIMPANDO DATASET
df_pm = df_pm.drop("IDADE_ALUNO_2020", 'ANOS_PM_2020', 'INDE_CONCEITO_2020', 'TURMA_2021', 'SINALIZADOR_INGRESSANTE_2021', 'TURMA_2022')

df_pm.show()
```

```python
#AJUSTANDO ORDEM DE COLUNAS
df_pm = df_pm.select("ID_ALUNO", "ANO_INGRESSO", "INSTITUICAO_ENSINO_ALUNO_2020", "FASE_TURMA_2020", "PEDRA_2020", "INDE_2020", "IAA_2020", "IEG_2020", 'DESTAQUE_IEG_2020', "IDA_2020", 'DESTAQUE_IDA_2020', "IPV_2020", 'DESTAQUE_IPV_2020', "IPP_2020", "IPS_2020", "IAN_2020", "PONTO_VIRADA_2020", "INSTITUICAO_ENSINO_ALUNO_2021", "FASE_2021", "PEDRA_2021", "INDE_2021", "IAA_2021", "IEG_2021", "IPS_2021", "IDA_2021", "IPP_2021", "IPV_2021", "IAN_2021", "REC_EQUIPE_1_2021", "REC_EQUIPE_2_2021", "REC_EQUIPE_3_2021", "REC_EQUIPE_4_2021", "PONTO_VIRADA_2021", "NIVEL_IDEAL_2021", "DEFASAGEM_2021", "BOLSISTA_2022", "FASE_2022", "PEDRA_2022", "INDE_2022", "IAA_2022", "IEG_2022", "DESTAQUE_IEG_2022", "IDA_2022", "DESTAQUE_IDA_2022", "IPV_2022", "DESTAQUE_IPV_2022", "IPS_2022", "IPP_2022", "IAN_2022", "CG_2022", "CF_2022", "CT_2022", "NOTA_PORT_2022", "NOTA_MAT_2022", "NOTA_ING_2022", "QTD_AVAL_2022", "REC_AVA_1_2022", "REC_AVA_2_2022", "REC_AVA_3_2022", "REC_AVA_4_2022", "INDICADO_BOLSA_2022", "PONTO_VIRADA_2022", "NIVEL_IDEAL_2022")

df_pm.show()
```

```python
#RENOMEANDO COLUNAS
df_pm = df_pm.withColumnRenamed("REC_AVA_1_2022", "REC_EQUIPE_1_2022").withColumnRenamed("REC_AVA_2_2022", "REC_EQUIPE_2_2022").withColumnRenamed("REC_AVA_3_2022", "REC_EQUIPE_3_2022").withColumnRenamed("REC_AVA_4_2022", "REC_EQUIPE_4_2022").withColumnRenamed("INSTITUICAO_ENSINO_ALUNO_2020", "IE_2020").withColumnRenamed("INSTITUICAO_ENSINO_ALUNO_2021", "IE_2021").withColumnRenamed("FASE_TURMA_2020", "FASE_2020")

df_pm.show()
```

```python
#TRANSF COLUNAS EM INT
df_pm = df_pm.withColumn("ANO_INGRESSO", floor(col("ANO_INGRESSO")).cast("integer"))
```
